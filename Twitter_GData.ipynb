```python
Search_words = "\u0645\u0631\u0648\u062C " # @param {type:"string"}
Date_since = "2014-01-01" # @param {type:"date"}
Count = 200 # @param {type:"integer"}
#-filter:quoteweets -filter:replies -filter:retweets

```

```python
tweets = tw.Cursor(api.search,
                       q=Search_words,
                       lang="ar",
                       since=Date_since).items(Count) 
                       
```

```python
users_tweet = [[tweet.user.screen_name, tweet.user.location, tweet.text, tweet.favorite_count, tweet.is_quote_status, tweet.created_at, tweet.retweet_count , ] for tweet in tweets]
```

```python
tweet_text = pd.DataFrame(data=users_tweet, 
                    columns=['user', "location","text","favorite_count","is_quote_status","created_at", "retweet_count"])


tweet_text
```
